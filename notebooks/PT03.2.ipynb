```python
# True species tree.
tree = "/home/henry/phylo-timescale/newick/realtree.tre"
truetree = toytree.tree(tree)
truetree.draw(layout='d', use_edge_lengths=True, node_labels="idx", node_sizes=12);
```

```python
# Set up bpp object.
bpp_ne = ipa.bpp(
    name="bpp_ne_small",
    data=loci,
    guidetree=truetree.drop_tips([i for i in truetree.get_tip_labels() if int(i[1:]) >5]), 
    imap={"r{}".format(i) : ["r{}".format(i)] for i in range(6)},
    maxloci=500,
    workdir="/home/henry/phylo-timescale/bpp",
    reps_resample_loci=False,
    )
```

```python
# Set parameters.
bpp_ne.kwargs['thetaprior'] = (3.5, 0.015)
bpp_ne.kwargs['tauprior'] = (8, 0.009)
bpp_ne.kwargs['sampfreq'] = 3
bpp_ne.kwargs['burnin'] = 100000
bpp_ne.kwargs['nsample'] = 1000000
```

```python
# Draw priors.
bpp_ne.draw_priors(
    gentime_min=0.8, 
    gentime_max=1.2,
    mutrate_min=1e-8,
    mutrate_max=1.001e-8,
);
```

```python
bpp_ne.ipcluster['cores'] = 8
bpp_ne.run(nreps=4, auto=True, force=True)
```

```python
res, mcmc = bpp_ne.summarize_results("00", individual_results = False)
```

```python
dfdiv, dfne, ttre, mtre = bpp_ne.transform(mcmc,
    gentime_min=0.8, 
    gentime_max=1.2,
    mutrate_min=1e-8,
    mutrate_max=1.001e-8,)
```
